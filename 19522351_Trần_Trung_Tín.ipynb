#dùng geopandas để đọc shapefile trong /Data/GIS/Population/population_HCMC/population_shapefile/Population_Ward_Level.shp

```python
import geopandas as gpd
```

```python
pop = gpd.read_file("/content/CSL_HCMC/Data/GIS/Population/population_HCMC/population_shapefile/Population_Ward_Level.shp")
pop
```

#hãy thực hiện các tác vụ truy vấn sau

```python
pop_sort = pop.sort_values(by="Shape_Area",ascending = False)
pop_sort = pop_sort.reset_index()
pop_sort.head(5)
```

```python
# Phường nào có diện tích lớn nhất
# Phường nào có diện tích nhỏ nhất

print("Phường có diện tích lớn nhất là phường {} quận {}.".format(*pop_sort.loc[0,["Com_Name","Dist_Name"]]))
index = pop_sort.loc[pop_sort.Shape_Area ==pop_sort.Shape_Area.min()].index[0]
print("Phường có diện tích nhỏ nhất là phường {} quận {}.".format(*pop_sort.loc[index,["Com_Name","Dist_Name"]]))
```

```python
pop_sort = pop.sort_values(by="Pop_2019",ascending = False)
pop_sort = pop_sort.reset_index()
pop_sort.head(5)
```

```python
# Phường nào có dân số 2019 (Pop_2019) cao nhất
# Phường nào có dân số thấp nhất (2019)

print("Phường có dân số cao nhất là phường {} quận {}.".format(*pop_sort.loc[0,["Com_Name","Dist_Name"]]))
index = pop_sort.loc[pop_sort.Pop_2019 ==pop_sort.Pop_2019.min()].index[0]
print("Phường có dân số nhỏ nhất là phường {} quận {}.".format(*pop_sort.loc[index,["Com_Name","Dist_Name"]]))
```

```python
pop["growth"] = pop["Pop_2019"] / pop["Pop_2009"]
pop_sort = pop.sort_values(by="growth",ascending = False)
pop_sort = pop_sort.reset_index()
pop_sort.head(5)
```

```python
# Phường nào có tốc độ tăng trưởng dân số nhanh nhất (dựa trên Pop_2009 và Pop_2019)
# Phường nào có tốc độ tăng trưởng dân số thấp nhất

print("Phường có tốc độ tăng trưởng dân số nhanh nhất là phường {} quận {}.".format(*pop_sort.loc[0,["Com_Name","Dist_Name"]]))
index = pop_sort.loc[pop_sort.growth ==pop_sort.growth.min()].index[0]
print("Phường có tốc độ tăng trưởng dân số chậm nhất là phường {} quận {}.".format(*pop_sort.loc[index,["Com_Name","Dist_Name"]]))
```

```python
pop["vol"] = pop["Pop_2019"] - pop["Pop_2009"]
pop_sort = pop.sort_values(by="vol",ascending = False)
pop_sort = pop_sort.reset_index()
pop_sort.head(5)
```

```python
# Phường nào có biến động dân số nhanh nhất
# Phường nào có biến động dân số chậm nhất

print("Phường có biến động dân số nhanh nhất là phường {} quận {}.".format(*pop_sort.loc[0,["Com_Name","Dist_Name"]]))
index = pop_sort.loc[pop_sort.vol ==pop_sort.vol.min()].index[0]
print("Phường có biến động dân số chậm nhất là phường {} quận {}.".format(*pop_sort.loc[index,["Com_Name","Dist_Name"]]))
```

```python
pop["den"] = pop["Pop_2019"] / pop.Shape_Area
pop_sort = pop.sort_values(by="den",ascending = False)
pop_sort = pop_sort.reset_index()
pop_sort.head(5)
```

```python
# Phường nào có mật độ dân số cao nhất (2019)
# Phường nào có mật độ dân số thấp nhất (2019)

print("Phường có mật độ dân số cao nhất là phường {} quận {}.".format(*pop_sort.loc[0,["Com_Name","Dist_Name"]]))
index = pop_sort.loc[pop_sort.den ==pop_sort.den.min()].index[0]
print("Phường có mật độ dân số thấp nhất là phường {} quận {}.".format(*pop_sort.loc[index,["Com_Name","Dist_Name"]]))
```
